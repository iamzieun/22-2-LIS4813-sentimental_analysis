# 데이터 불러오기

```python
df_raw = pd.read_csv("/content/drive/MyDrive/[22-2] 딥러닝개론/project/소상공인시장진흥공단_상가(상권)정보_서울_202206.csv")
df_raw.head()
```

# 행 선택

## 상권업종대분류명 / 중분류명

```python
# 대분류명
df_raw["상권업종대분류명"].unique() # 음식
df_category_1 = df_raw[df_raw["상권업종대분류명"] == "음식"]
df_category_1.head()
```

```python
# 중분류명
df_category_1["상권업종중분류명"].unique()
```

```python
# 중분류명이 '커피점/카페', '제과제빵떡케익'인 행 drop
df_category_2 = df_category_1[(df_category_1["상권업종중분류명"] != "커피점/카페") & (df_category_1["상권업종중분류명"] != "제과제빵떡케익")]
df_category_2.head()
```

## 시도명 / 시군구명

```python
df_category_2["시도명"].unique()    # 서울특별시
df_category_2["시군구명"].unique()  # 서대문구
```

```python
df_region = df_category_2[df_category_2["시군구명"] == "서대문구"]
df_region.head()
```

```python
df_region.shape # 총 2844개 음식점
```

## 이름 열 만들기 (상호명 + 지점명)

```python
# 인덱스 초기화
df_region.index = range(df_region.shape[0])
df_region.head()
```

```python
# 지점명 열 결측치 빈칸으로 채우기
df_region["지점명"].fillna("", inplace=True)
```

```python
# 이름 열 만들기
df_region["이름"] = df_region["상호명"] + " " + df_region["지점명"]
df_region
```

```python
# 이름 열 결측치 확인
df_region["이름"].isnull().sum()
```

```python
# 행만 정제한 데이터프레임 저장
df_row_cleared = df_region
df_row_cleared.to_csv("/content/drive/MyDrive/[22-2] 딥러닝개론/project/publicdata_rowcleared.csv", sep=',')
```
