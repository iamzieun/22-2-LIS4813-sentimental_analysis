**필요한 것: 리뷰문장, 이중분류라벨(0,1), 삼중분류라벨(0,1,2)**

```python
# rev1
# 이중분류는 이미 있음, 0,1,2
for i in range(0,len(rev1)):
    if rev1.loc[i,'score']<3:
        rev1.loc[i,'trinary']=0
    elif rev1.loc[i,'score']==3:
        rev1.loc[i,'trinary']=1
    else:
        rev1.loc[i,'trinary']=2
        
rev1.rename(columns = {'y' : 'binary'}, inplace = True)
rev1
```

```python
# rev2
# 이중분류는 이미 있음.
for i in range(0,len(rev2)):
    if rev2.loc[i,'score']<3:
        rev2.loc[i,'trinary']=0
    elif rev2.loc[i,'score']==3:
        rev2.loc[i,'trinary']=1
    else:
        rev2.loc[i,'trinary']=2

rev2['trinary']=rev2['trinary'].astype(int)
rev2.rename(columns = {'y' : 'binary'}, inplace = True)
rev2
```

```python
#rev3
#이중분류
rev3.rename(columns = {'리뷰':'review','평점' : 'score'}, inplace = True)

for i in range(0,len(rev3)):
    if rev3.loc[i,'score']<4:
        rev3.loc[i,'binary']=0
    else:
        rev3.loc[i,'binary']=1
        
# 삼중분류
for i in range(0,len(rev3)):
    if rev3.loc[i,'score']<3:
        rev3.loc[i,'trinary']=0
    elif rev3.loc[i,'score']==3:
        rev3.loc[i,'trinary']=1
    else:
        rev3.loc[i,'trinary']=2
        
rev3=rev3.astype({'binary':'int','trinary':'int'})
rev3
```

```python
# 필요한 열만 추출하고 합치기
rev1_ = rev1[['review','score','binary','trinary']]
rev2_ = rev2[['review','score','binary','trinary']]
rev3_ = rev3[['review','score','binary','trinary']]

rev_data = pd.concat([rev1_,rev2_,rev3_])
rev_data
```

```python
rev_data.to_csv('/Users/chaesung/Desktop/딥러닝개론 프로젝트/traindata/rev_data.csv')
```
