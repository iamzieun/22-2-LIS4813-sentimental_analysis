# Data preprocessing

```python
total_review = pd.DataFrame(columns=['식당 인덱스', '리뷰'])
total_review
```

```python
num_idx = 0

for i in range(len(total_info)):
    for j in total_info.loc[i, '리뷰'][2:-2].split("', '"):
        total_review.loc[num_idx, '식당 인덱스'] = i
        total_review.loc[num_idx, '리뷰'] = j
        num_idx += 1
```
